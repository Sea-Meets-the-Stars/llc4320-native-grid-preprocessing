```python
endpoint_url = "https://mghp.osn.xsede.org"

fs = s3fs.S3FileSystem(
    anon=True,
    client_kwargs={"endpoint_url": endpoint_url},
)

files = fs.ls("cnh-bucket-1/llc_surf/kerchunk_files/")
```

```python
files
```

```python
import re

prefix = "llc4320_Eta-U-V-W-Theta-Salt_f"
pattern = re.compile(r"^llc4320_Eta-U-V-W-Theta-Salt_f\d+_k\d+_iter_(\d+)\.json$")

max_file = None
max_iter = -1
theta_files = []
for f in files:
    name = f.split("/")[-1]
    if not name.startswith(prefix):
        continue

    m = pattern.match(name)
    if m:
        theta_files.append(f)
        it = int(m.group(1))
        if it > max_iter:
            max_iter = it
            max_file = f

print("max_iter:", max_iter)
print("max_file:", max_file)
```
